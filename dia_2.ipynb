# Criptografia
## Hash

Un hash es un algoritmo matemático que toma una entrada (un mensaje, archivo, contraseña u otro dato) y lo procesa para producir una salida única, de longitud fija y aparentemente aleatoria. Esta salida, a menudo llamada "hash" o "resumen criptográfico", es esencialmente una representación numérica del contenido de la entrada.

```python
#Importar la libreria sha256
from hashlib import sha256

# Creamos un mensaje
mensaje = 'Hola pinguinas'

# Hasheamos el mensaje
mensaje_hasheado = sha256(mensaje.encode()).hexdigest()

# Imprimimos
print(mensaje_hasheado)

```

# Implementacion del hash en la blockchain

```python
from datetime import datetime

#importamos sha
from hashlib import sha256
# Importamos Json
import json

tiempo = str(datetime.today())
class Bloque:  

    # 2 Agregamos el hash anterior en la inicializacion
    def __init__(self, index, transaccion, timestamp, hash_anterior):
        self.index = index
        self.transaccion = transaccion
        self.timestamp = timestamp

        self.hash_anterior = hash_anterior

class Blockchain:
    def __init__(self):
        self.transacciones_pendientes = []

        self.cadena = []

        self.tiempo = tiempo

        self.crear_bloque_genesis()
    
    # 3 Agregamos el hash 0 al bloque genesis
    def crear_bloque_genesis(self):
        bloque_genesis = Bloque(0, ['Bloque genesis'], self.tiempo, '0')

        self.cadena.append(bloque_genesis)

    def agregar_bloque(self, bloque):
        self.cadena.append(bloque)
        return True
    
    def agregar_transaccion(self, transaccion):
        self.transacciones_pendientes.append(transaccion)

    @property
    def ultimo_bloque(self):
        return self.cadena[-1]
    

    # 1 Crear un metodo de hasheo que cripte los bloques

    # #El decorador @staticmethod se utiliza para declarar un método de clase como un método estático. Un método estático es un método que pertenece a la clase en lugar de una instancia de la clase. Esto significa que no se necesita una instancia de la clase para llamar al método estático.
    @staticmethod
    def criptar(bloque):
        bloque_string = json.dumps(bloque.__dict__, sort_keys=True)

        return sha256(bloque_string.encode()).hexdigest()
    

    # 4 Agregamos el hash del bloque anterior al bloque que se va a cerrar
    def cerrar_bloque(self):
        if not self.transacciones_pendientes:
            return False
        ultimo_bloque = self.ultimo_bloque


        # Llamamos a nuestra funcion criptar
        nuevo_bloque = Bloque(ultimo_bloque.index + 1, self.transacciones_pendientes, self.tiempo, self.criptar(ultimo_bloque))

        self.agregar_bloque(nuevo_bloque)

        self.transacciones_pendientes = []
    
        return nuevo_bloque.index




pengucoin = Blockchain()

# Agregamos 2 transacciones
pengucoin.agregar_transaccion('cualquier cosa')
pengucoin.agregar_transaccion('otra cosa')

# Cerramos el bloque
pengucoin.cerrar_bloque()
pengucoin.agregar_transaccion('tercera cosa')
pengucoin.cerrar_bloque()


print('Bloque Genesis')
print(pengucoin.cadena[0].__dict__)

# Hasheamos el bloque imprimido
hash_del_bloque = pengucoin.criptar(pengucoin.cadena[0])
print(hash_del_bloque)

print('------------------')


print('Primer bloque')
print(pengucoin.cadena[1].__dict__)

# Hasheamos el bloque imprimido
hash_del_bloque = pengucoin.criptar(pengucoin.cadena[1])
print(hash_del_bloque)

print('------------------')

print('Segundo bloque')
print(pengucoin.cadena[2].__dict__)

# Hasheamos el bloque imprimido
hash_del_bloque = pengucoin.criptar(pengucoin.cadena[2])
print(hash_del_bloque)


    
```

# Proof of Work (Prueba de trabajo)

Para validar las transacciones y crear nuevos bloques, se utiliza un algoritmo llamado "Proof of Work" (PoW), que es un sistema que requiere que los mineros resuelvan un problema matemático muy complejo para encontrar un número aleatorio conocido como "nonce". Este nonce se agrega al bloque junto con las transacciones, y luego se aplica una función hash criptográfica para producir un valor hash que debe cumplir con ciertos requisitos.

# Nonce (Number used once)

El objetivo del nonce en la blockchain es encontrar un valor hash del bloque que cumpla con una condición específica, también conocida como "dificultad" de la minería. El minero utiliza el nonce para generar diferentes valores de hash hasta que encuentre uno que cumpla con la dificultad requerida. El proceso de encontrar el valor hash correcto implica una gran cantidad de cálculos matemáticos y es lo que se conoce como minería de criptomonedas.


# Ejemplo de Nonce en python

```python
from hashlib import sha256

mensaje300 = 'Hola, este es mi privado'

# Hasheamos
mensaje_hasheadito = sha256(mensaje300.encode()).hexdigest()

# Numero de iteraciones que se haran para encontrar el hash
iteraciones = 0

# Creamos un nonce
nonce = 0


while True:
    #Incrementa el nonce en cada iteracion
    nonce += 1

    # Unir el mensaje con nuestro actual

    mensaje_con_nonce = f'{mensaje_hasheadito} {nonce}'

    # Calcular el hash

    hash_actual = sha256(mensaje_con_nonce.encode()).hexdigest()
    print(nonce)
    if hash_actual.startswith('0000'):
        print(f'Hash encontrado {hash_actual}')
        break


print(f'Nonce encontrado: {nonce}')
```

# Implementacion a la blockchain

```python
from datetime import datetime
from hashlib import sha256
import json

tiempo = str(datetime.today())


class Bloque:
    # 1 Agregar el atributo nonce  
    def __init__(self, index, transaccion, timestamp, hash_anterior, nonce=0):
        self.index = index
        self.transaccion = transaccion
        self.timestamp = timestamp
        self.hash_anterior = hash_anterior
        # Agregamos el atributo nonce
        self.nonce = nonce

    # 2 Creamos el metodo de criptar
    def criptador(self):
        bloque_string = json.dumps(self.__dict__, sort_keys=True)

        return sha256(bloque_string.encode()).hexdigest()

class Blockchain:
    def __init__(self):
        self.transacciones_pendientes = []
        self.cadena = []
        self.tiempo = tiempo
        self.crear_bloque_genesis()
    
    def crear_bloque_genesis(self):
        bloque_genesis = Bloque(0, ['Bloque genesis'], self.tiempo, '0')

        # Encriptamos el bloque genesis
        bloque_genesis.hash = bloque_genesis.criptador()
        self.cadena.append(bloque_genesis)

    @property
    def ultimo_bloque(self):
        return self.cadena[-1]   



    # Paso 3
    # Agregamos la dificultar a la blockchain
    dificultad = 3

    # Creamos la funcion de prueba de trabajo (Proof of work)
    def prueba_trabajo(self, bloque):
        # Comenzamos con el nonce 0
        bloque.nonce = 0

        # Criptamos el bloque para ver el hash
        bloque_criptado = bloque.criptador()

        # Comparar el hash del bloque segun el ajuste de la dificultad
        while not bloque_criptado.startswith('0'* self.dificultad):
            # Vamos incrementando hasta que el hash comience con la cantidad requerida
            bloque.nonce += 1

            # Criptamos el bloque con el nuevo valor del hash
            bloque_criptado = bloque.criptador()
        return bloque_criptado
    
    # Paso 4
    # Creamos una funcion que devuelva la validez del bloque
    # verificando si el hash correcto
    def prueba_de_trabajo_validada(self, bloque, hash_del_bloque):
        return(hash_del_bloque.startswith('0'* Blockchain.dificultad) and 
               hash_del_bloque == bloque.criptador())
    

    # Paso 5
    # Agregamos la validacion al agregar el bloque
    def agregar_bloque(self, bloque, prueba):

        # Traemos el hash anterior
        hash_anterior = self.ultimo_bloque.hash

        # Si el hash es diferente al hash del bloque anterior que nos devuelva falso
        if hash_anterior != bloque.hash_anterior:
            return False
        
        # Si no se valida, que nos retorne falso tambien
        if not self.prueba_de_trabajo_validada(bloque, prueba):
            return False
        
        # Guardamos la prueba en el hash del bloque
        bloque.hash = prueba


        self.cadena.append(bloque)
        return True
    
    def agregar_transaccion(self, transaccion):
        self.transacciones_pendientes.append(transaccion)

    # Por ultimo hacemos el
    # Paso 7

    def cerrar_bloque(self):
        if not self.transacciones_pendientes:
            return False
        ultimo_bloque = self.ultimo_bloque

        # definimos cada parametro y modificamos para que nos traiga el ultimo hash
        nuevo_bloque = Bloque(ultimo_bloque.index + 1, transaccion=self.transacciones_pendientes, timestamp=tiempo, hash_anterior=ultimo_bloque.hash)
        
        # Agregamos la prueba que hicimos anteriormente
        prueba = self.prueba_trabajo(nuevo_bloque)

        #Al agregar el nuevo bloque tambien tenemos que entregarle la prueba para confirmar que sigue la cadena
        self.agregar_bloque(nuevo_bloque, prueba)

        self.transacciones_pendientes = []
        return nuevo_bloque.index


# Les dejo este ejemplo extra por si quieran hacer la prueba
# En jupiter o desde python


# Instanciamos el bloque
penguincoin = Blockchain()

#Agregamos transacciones
penguincoin.agregar_transaccion("Transaccion 1")
penguincoin.agregar_transaccion("Transaccion 2")

#Minamos el bloque 
penguincoin.cerrar_bloque()

#Imprimimos el bloque
for bloque in penguincoin.cadena:
    print(bloque.__dict__)
```
